Loading the combined file and list the column names

```python
from pathlib import Path
import pandas as pd

# 1. Point to your combined CSV file
file_path = Path("combined_momo_statements.csv")

# 2. Load it into a DataFrame
df = pd.read_csv(file_path)

# 3. Print the initial setup to see what we are starting with
print(f"File loaded successfully!")
print(f"Initial Dimensions: {df.shape[0]} rows and {df.shape[1]} columns.\n")
print("Columns currently in the file:")
print(list(df.columns))
```

Droping empty columns into our new file

```python
# 1. Drop columns where ALL values are missing (NaN)
# axis=1 targets columns; how='all' ensures we only drop 100% empty ones
df_clean = df.dropna(axis=1, how="all")

# 2. Calculate and display exactly what changed
removed_cols = set(df.columns) - set(df_clean.columns)

print("--- Cleaning Summary ---")
print(f"Columns removed ({len(removed_cols)}): {list(removed_cols)}")
print(f"Remaining columns ({len(df_clean.columns)}): {list(df_clean.columns)}")
print(f"New Dimensions: {df_clean.shape[0]} rows and {df_clean.shape[1]} columns.\n")

# 3. Preview the cleaned data
df_clean.head()
```

Reviewing the cleaning dataset

```python
# 1. Define the name for your newly cleaned dataset
clean_output_file = Path("cleaned_momo_statements.csv")

# 2. Save the cleaned DataFrame back to your current folder
# index=False ensures Pandas doesn't add an unnecessary row-number column
df_clean.to_csv(clean_output_file, index=False)

print(f"🎉 Success! The cleaned dataset has been saved as: '{clean_output_file.name}'")
print("You can now open this file in your folder to review the clean layout.")
```

```python
from pathlib import Path
import pandas as pd

# 1. Point to your clean file
clean_file = Path("cleaned_momo_statements.csv")

# 2. Load it into a new DataFrame
df_dates = pd.read_csv(clean_file)

print("File loaded successfully!")
print(f"Current columns available: {list(df_dates.columns)}")

# Let's peek at how the date column looks right now
# Note: Replace 'Date' with the exact name of your column if it's different (e.g., 'date', 'Timestamp')
date_col_name = 'Date' 
print(f"\nCurrent data type of '{date_col_name}': {df_dates[date_col_name].dtype}")
print("First 3 date samples look like this:")
print(df_dates[date_col_name].head(3))
```

Cell 2: Convert the date column to the right data typeConvert the date column to the right data type

```python
# 1. Define your date column name
date_col_name = 'Date'

# 2. Convert the column to datetime
# dayfirst=True is highly recommended if your dates are formatted like DD/MM/YYYY
df_dates[date_col_name] = pd.to_datetime(df_dates[date_col_name], dayfirst=True)

print(f"✅ Success! New data type of '{date_col_name}': {df_dates[date_col_name].dtype}")
print(f"Earliest date found in file: {df_dates[date_col_name].min()}")
print(f"Latest date found in file: {df_dates[date_col_name].max()}")
```

```python
# 1. Define your start and end targets
start_date = "2026-01-01"
end_date = "2026-04-30"

# 2. Filter the rows using a boolean mask
# This keeps rows where the date is greater than or equal to Jan 1 AND less than or equal to Apr 30
date_filter = (df_dates[date_col_name] >= start_date) & (df_dates[date_col_name] <= end_date)
df_filtered = df_dates[date_filter]

# 3. Sort by date so it's in chronological order
df_filtered = df_filtered.sort_values(by=date_col_name)

print("--- Filtering Summary ---")
print(f"Rows before date filter: {df_dates.shape[0]}")
print(f"Rows after filtering (Jan-Apr 2026): {df_filtered.shape[0]}")
print(f"New date boundaries: {df_filtered[date_col_name].min()} to {df_filtered[date_col_name].max()}")

# Preview the final filtered data
df_filtered.head()
```
